## Распределение Рэлея

### формулы для рассчетов
#### Математическое ожидаие : $ \sqrt{ \frac{\pi}{2} }\sigma $
#### Дисперсия : $ (2 - \frac{\pi}{2}) \sigma^2 $
#### Теоретическая плотность распределения :  $ \frac {x}{\sigma^2}\exp(- \frac{x^2}{2\sigma^2})$

```python
import math 
sigma = 1
#рассчитываю математическое ожидание и дисперсию по формулам выше
M = math.sqrt(math.pi / 2) * sigma
D = (2 - math.pi / 2) * math.pow(sigma, 2)
print M, D
```

Генерирую выборку объёма 1000, строю гистограмму выборки и рисую поверх неё теоретическую плотность распределения случайной величины

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
r = sts.rayleigh(scale = sigma)
r1 = r.rvs(size=1000)
plt.hist(r1,bins=20, normed=True, color = "blue", label = 'Sampling')
std = math.sqrt(D)
#задаю теоретическую плотность распределения
def f(x):
    return (x/sigma**2) * np.exp(- (x**2 / 2.) * sigma**2)

x = np.linspace(0.,5.,1000)
func = f(x)
plt.plot(x, func, color="red", label = 'Probability density' )

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("n = 1")
plt.legend()
plt.show()
```

генерирую 1000 выборок объемом 5

```python
meanAr = []
a = np.array([])
r = sts.rayleigh(scale = 1)
for i in range(1000):
    r2 = r.rvs(size=5)
    a=np.append(a, r2.mean()) #Нахожу среднее значение и записываю в массив

#Построение гистограммы средних значений
plt.hist(a,bins=20,  normed=True, color = "blue", label = 'Std')


sig5 = (D / 5.)**0.5 #нахождение второго аргумент для нормального распределение по цпт
norm_rv = sts.norm(M, sig5)
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, color = 'red',  label='Probability density')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("n = 5")
plt.legend()
plt.show()
```

Генерирую 1000 выборок объемом 10

```python
r = sts.rayleigh(scale = 1)
a10 = np.array([])
for i in range(1000):
    r3 = r.rvs(size=10)
    a10=np.append(a10, r3.mean())
print type(a10)

sig10 = (D / 10.)**0.5
norm_rv = sts.norm(M, sig10)
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, color = 'red',  label='Probability density')
plt.hist(a10,bins=20, normed=True, color = "blue", label = 'Std')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("n = 10")
plt.legend()
plt.show()
```
