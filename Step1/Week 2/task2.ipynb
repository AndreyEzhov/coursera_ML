```python
import numpy
from scipy import linalg
import math 
%matplotlib inline
```

```python
def calculate_f_x(x):
    return numpy.sin(x / 5) * numpy.exp(x / 10) + 5 * numpy.exp(-x / 2)

a = numpy.array([[1,1],[1,15]])
b = numpy.array([calculate_f_x(1.0), calculate_f_x(15.0)])
print linalg.solve(a,b)
```

```python
a = numpy.array([[1,1,1,1],[1,4 ** 1, 4 ** 2, 4 ** 3],[1,10 ** 1, 10 ** 2, 10 ** 3],[1,15 ** 1, 15 ** 2, 15 ** 3]])
b = numpy.array([calculate_f_x(1.0), calculate_f_x(4.0), calculate_f_x(10.0), calculate_f_x(15.0)])
result = linalg.solve(a,b)
```

```python
import matplotlib.pyplot as plt
```

```python
x = numpy.arange(1, 15, 0.1)
y = result[0] + result[1] * x + result[2] * x ** 2 + result[3] * x ** 3
plt.plot(x, y)
y_o = calculate_f_x(x)
plt.plot(x, y_o)
plt.show()
```
